```python
import pandas as pd

df = pd.read_csv("valid.csv")

df_out = (
    df.pivot_table(
        index=["text", "variety"],
        columns="task",
        values="label",
        aggfunc="first"
    )
    .reset_index()
)

df_out.rename(columns={"Sentiment": "sentiment", "Sarcasm": "sarcasm"}, inplace=True)
df_out =df_out.dropna()
df_out['sarcasm'] = df_out['sarcasm'].fillna(0).astype(int)
df_out['sentiment'] = df_out['sentiment'].fillna(0).astype(int)

df_out.to_csv("output.csv", index=False)
```
