```python
import pandas as pd

def merge_tasks_with_nan(splice):

    df = pd.read_csv(f"{splice}.csv")

    df_out = (
        df.pivot_table(
            index=["text", "variety", "source"],
            columns="task",
            values="label",
            aggfunc="first"
        )
        .reset_index()
    )
    print(df_out.head())

    df_out.rename(columns={"Sentiment": "sentiment", "Sarcasm": "sarcasm"}, inplace=True)
    # df_out =df_out.dropna()

    df_out['sarcasm'] = df_out['sarcasm'].astype('Int64')
    df_out['sentiment'] = df_out['sentiment'].astype('Int64')
    # df_out['sarcasm'] = df_out['sarcasm'].fillna(0).astype(int)
    # df_out['sentiment'] = df_out['sentiment'].fillna(0).astype(int)

    df_out.to_csv(f"{splice}_SS_with_nan.csv", index=False)

merge_tasks_with_nan("train")
merge_tasks_with_nan("valid")
```
